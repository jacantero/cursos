```
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0,10,100)
Y = np.sin(X)
plt.plot(X,Y)
Z = np.cos(X)
plt.plot(X,Z)
plt.legend(['Sinus','Cosinus'])
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,50,1)
y = 3*x
plt.plot(x,y)
plt.title("Draw a line")
plt.xlabel("x - axis")
plt.xlim(0,50)
plt.ylabel("y - axis")
plt.ylim(0,160)
plt.grid('on')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,3,5)
y = np.zeros(len(x))
for i in range (0,len(x)):
  if i<=2:
    y[i] = 2*x[i]
  else:
    y[i] = 10 - 3*x[i]
print(x)
print(y)
plt.plot(x,y)
plt.xlabel("x - axis")
plt.xlim(1,3)
plt.ylabel("y - axis")
plt.ylim(1,4)
plt.title("Triangle function")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10,31,10)
y1 = np.array([2*x[i] if x[i]<=20 else 100 - 3*x[i] for i in range(0,len(x))])
y2 = np.array([70 - 3*x[i] if x[i]<=20 else -30 + 2*x[i] for i in range(0,len(x))])
print(x,y1,y2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x - axis")
plt.ylabel("y - label")
plt.legend("Double triangle plot")
plt.xlim(10,30)
plt.ylim(10,40)
plt.legend(['y1','y2'])
plt.show()
```
