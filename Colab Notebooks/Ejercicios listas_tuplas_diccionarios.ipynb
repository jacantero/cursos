```
salir = False
contactos = {}
while (not salir):
  print("Introduce el nombre del usuario:")
  x = input()
  if x in contactos:
    print("Este contacto ya existe. Por favor, cambia su nombre e inténtalo de nuevo.")
    x = input()
  print("Introduce el nº de teléfono del usuario:")
  y = input()
  if x not in contactos:
    contactos[x] = y
  else:
    print("El contacto introducido ya existe. Por favor, cambia su nombre e inténtalo de nuevo.")
  print("¿Quieres seguir introduciendo nuevos contactos?")
  z=input()
  if z == "No":
    salir = True
print(contactos)
```
